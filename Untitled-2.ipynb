```python
import pandas as pd
```

```python
data_dir = './data/trial1/'
```

```python
df = pd.read_csv(data_dir+'driving_log.csv', names=['center_cam', 'left_cam', 'right_cam', 'steering', 'throttle', 'brake', 'speed'])
```

```python
df.head()
```

```python
df.info()
```

```python
df_clean = df.copy()
```

```python
df_clean.iloc[:,0:3] = df_clean.iloc[:,0:3].apply(lambda x: data_dir + 'IMG/' + x.str.split('\\').str[-1])
```

```python
df_clean.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
images = []
measurements = []
```

```python
df_clean['center_cam'].apply(lambda x: images.append(mpimg.imread(x)));
df_clean['steering'].apply(lambda x: measurements.append(x));
```

```python
import numpy as np
```

```python
X_train = np.array(images)
y_train = np.array(measurements)
```

# Training

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Flatten(input_shape=(160,320,3)))
model.add(Dense(1));
```

```python
model.compile(loss='mse', optimizer='adam')
model.fit(X_train, y_train, epochs=7, validation_split=0.2, shuffle=True)
```

```python
model.save('model.h5')
```

```python
from tensorflow.keras.models import load_model
```
